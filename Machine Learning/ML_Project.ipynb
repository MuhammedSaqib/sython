```python
import pandas as pd
import numpy as np
```

```python
trainFN=pd.read_csv('train_FNC.csv')
trainSB=pd.read_csv('train_SBM.csv')

y=pd.read_csv('train_labels.csv')

testFN=pd.read_csv('test_FNC.csv')
testSB=pd.read_csv('test_SBM.csv')

ty=pd.read_csv("submission_example.csv")
```

```python
train= pd.merge(trainFN,trainSB, on='Id')
test= pd.merge(testFN,testSB, on="Id")

a=y.drop('Id',axis=1)

traina=train.drop('Id', axis=1)
testa=test.drop('Id', axis=1)
train.shape
test.shape
train.head()
print train.shape
print y.shape
print a.shape
print traina.shape
```

```python
from sklearn.cross_validation import cross_val_score
from sklearn.ensemble import RandomForestClassifier

clf = RandomForestClassifier(n_estimators=1000,max_features=50)

a=numpy.ravel(a)


clf.fit(traina, a)

output = clf.predict(testa)
```

```python
scores = cross_val_score(clf, traina, a)
scores.mean() 

#output = clf.predict(test)
```

```python
from sklearn import svm
import numpy

clf = svm.SVC()
a=numpy.ravel(a)
clf.fit(traina, a)

output = clf.predict(testa)

srce=clf.score(traina, a)
srce.mean()
```
